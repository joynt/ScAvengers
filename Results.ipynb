# LSTM part

```python
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
n_steps = 50
tot_X = np.zeros((1,n_steps), dtype=int)
tot_y = np.zeros((1,1), dtype=int)

c_to_use = clusters[cluster] # cambiare questo valore a seconda del cluster

for i in range(len(c_to_use)):
    dataset = c_to_use[i]
    dataset = dataset[:,4]
    X, y = split_sequence(dataset, n_steps=n_steps)
    tot_X = np.vstack((tot_X, X))
    y = y.reshape((y.shape[0],1))
    tot_y = np.vstack((tot_y, y))
    
print(tot_X.shape, tot_y.shape)
```

```python
final_dataset = np.hstack((tot_X, tot_y))
del tot_X, tot_y

np.random.seed(1)
for i in range(perm_times[cluster]):
    perm = np.random.permutation(len(final_dataset))
final_dataset = final_dataset[perm]
```

```python
# normalize dataset
scaler = MinMaxScaler(feature_range=(-1,1))
final_dataset = scaler.fit_transform(final_dataset)
```

```python
n_train = int(final_dataset.shape[0]*0.7)
#n_test = final_dataset.shape[0]-n_train 

X_train = final_dataset[:n_train,:-1]
y_train = final_dataset[:n_train,0]
X_test = final_dataset[n_train:,:-1]
y_test = final_dataset[n_train:,0]

print('Train length: {}\nTest length {}'.format(len(y_train),len(y_test)))
```

```python
# define the model
#n_batch = 1024
model = Sequential()
model.add(LSTM(100, input_shape=(n_steps, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mae')
```

# Compute error vector

```python
user_sep = []

for i in range(len(c_to_use)):
    user_sep.append(c_to_use[i].shape[0]-n_steps + 1)
```

```python
def compute_sliding_error_vector(pred, target, stride=1, window_size=20):
    
    error_list = []
    idx = 0
    
    while idx+window_size<target.shape[0]:
        distance = np.mean(np.abs(pred[idx:idx+window_size]-target[idx:idx+window_size]))
        error_list.append(distance)
        idx += stride
        
    return np.array(error_list)
```

```python
er_vector = compute_sliding_error_vector(y_test_pred_rescaled, y_test_rescaled)
```

```python
err_vec_separated = []
y_test_rescaled_s = []
y_test_pred_rescaled_s = []
start_idx = 0

for elem in user_sep:
    err_vec_separated.append(er_vector[start_idx:start_idx+elem])
    y_test_rescaled_s.append(y_test_rescaled[start_idx:start_idx+elem])
    y_test_pred_rescaled_s.append(y_test_pred_rescaled[start_idx:start_idx+elem])
    start_idx += elem

    
```

```python
err_vec_separated = [i for i in err_vec_separated if i.shape[0] > 0]
```

```python
errors = []
for count, err_vec in enumerate(err_vec_separated):
    error = np.zeros(err_vec.shape[0])
    mean = np.mean(err_vec)
    var = np.var(err_vec)
    index = err_vec > (mean + var)
    error[index] = 1
    errors.append(error)
```

```python
def smooth(error, th_in = 5, th_out = 5):
    window_in = list(error[:th_in])
    window_out = list(error[:th_out])
    smoothed = []
    state = 0
    to_in = len(error)
    to_out = len(error)
    if th_in > th_out:
        to_out -= th_in - th_out
    elif th_out > th_in:
        to_in -= th_out - th_in
    
    for value_in, value_out in zip(error[th_in:to_in], error[th_out:to_out]):
        if sum(window_in) == th_in:
            state = 1
        elif sum(window_out) == 0:
            state = 0
        smoothed.append(state)
        window_in.pop(0)
        window_out.pop(0)
        window_in.append(value_in)
        window_out.append(value_out)
    return np.array(smoothed)

def get_absd(vector):
    x = np.array(range(1,vector.shape[0]+1))
    dxdy = diff(vector)/diff(x)
    return np.abs(dxdy)
```

```python
index = 0
plt.rcParams.update({'font.size': 17})
x = smooth(errors[index], th_in = 20, th_out = 50)*8-1
fig, (ax1, ax2) = plt.subplots(2,1,sharex=True, figsize = (15,10))
ax1.set_xlim(0,len(x))
ax1.set_ylim(0,31)

ax1.set_ylabel('MCS')
ax1.plot(y_test_rescaled_s[index], label = 'Ground truth', linewidth = .7)
ax1.plot(x*6)
ax1.plot(y_test_pred_rescaled_s[index],  '--',label = 'Predicted', linewidth = .7)




ax1.legend()


ax2.set_ylim(0,6)

ax2.hlines([np.mean(err_vec_separated[index]), np.mean(err_vec_separated[index])+np.var(err_vec_separated[index])], xmin = 0, xmax = len(err_vec_separated[0]))
ax2.plot(err_vec_separated[index])
ax2.plot(x)
ax2.set_xlabel('Time [arrivals]')
ax2.set_ylabel('Error')
ax2.fill_between(np.arange(len(x)), 0, x, where=x > 0, facecolor='orange', alpha=0.5)
ax1.fill_between(np.arange(len(x)), 0, x*6, where=x > 0, facecolor='orange', alpha=0.5)
```

# Statistical description

```python
import scipy.stats as st

std1 = np.zeros((len(c1,)))
std2 = np.zeros((len(c2,)))
std3 = np.zeros((len(c3,)))
mean1 = np.zeros((len(c1,)))
mean2 = np.zeros((len(c2,)))
mean3 = np.zeros((len(c3,)))
iqr1 = np.zeros((len(c1,)))
iqr2 = np.zeros((len(c2,)))
iqr3 = np.zeros((len(c3,)))
variation1 = np.zeros((len(c1,)))
variation2 = np.zeros((len(c2,)))
variation3 = np.zeros((len(c3,)))

for i in range(len(c1)):
    std1[i] = np.std(c1[i][:,4])
    mean1[i] = np.mean(c1[i][:,4])
    iqr1[i] = st.iqr(c1[i][:,4])
    variation1[i] = st.variation(c1[i][:,4])
for i in range(len(c2)):
    std2[i] = np.std(c2[i][:,4])
    mean2[i] = np.mean(c2[i][:,4])
    iqr2[i] = st.iqr(c2[i][:,4])
    variation2[i] = st.variation(c2[i][:,4])
for i in range(len(c3)):
    std3[i] = np.std(c3[i][:,4])
    mean3[i] = np.mean(c3[i][:,4])
    iqr3[i] = st.iqr(c3[i][:,4])
    variation3[i] = st.variation(c3[i][:,4])
```

```python
mean_std_c1, err_std_c1 = (np.mean(std1), np.std(std1))
mean_std_c2, err_std_c2 = (np.mean(std2), np.std(std2))
mean_std_c3, err_std_c3 = (np.mean(std3), np.std(std3))
mean_avg_c1, err_avg_c1 = (np.mean(mean1), np.std(mean1))
mean_avg_c2, err_avg_c2 = (np.mean(mean2), np.std(mean2))
mean_avg_c3, err_avg_c3 = (np.mean(mean3), np.std(mean3))
mean_iqr_c1, err_iqr_c1 = (np.mean(iqr1), np.std(iqr1))
mean_iqr_c2, err_iqr_c2 = (np.mean(iqr2), np.std(iqr2))
mean_iqr_c3, err_iqr_c3 = (np.mean(iqr3), np.std(iqr3))
mean_variation_c1, err_variation_c1 = (np.mean(variation1), np.std(variation1))
mean_variation_c2, err_variation_c2 = (np.mean(variation2), np.std(variation2))
mean_variation_c3, err_variation_c3 = (np.mean(variation3), np.std(variation3))
```

```python
print('\nMean:')
print('Cluster 1:', mean_avg_c1, '±', err_avg_c1)
print('Cluster 2:', mean_avg_c2, '±', err_avg_c2)
print('Cluster 3:', mean_avg_c3, '±', err_avg_c3)
print('\nStandard deviation:')
print('Cluster 1:', mean_std_c1, '±', err_std_c1)
print('Cluster 2:', mean_std_c2, '±', err_std_c2)
print('Cluster 3:', mean_std_c3, '±', err_std_c3)
print('\nInterquartile:')
print('Cluster 1:', mean_iqr_c1, '±', err_iqr_c1)
print('Cluster 2:', mean_iqr_c2, '±', err_iqr_c2)
print('Cluster 3:', mean_iqr_c3, '±', err_iqr_c3)
print('\nVariation:')
print('Cluster 1:', mean_variation_c1, '±', err_variation_c1)
print('Cluster 2:', mean_variation_c2, '±', err_variation_c2)
print('Cluster 3:', mean_variation_c3, '±', err_variation_c3)
```
